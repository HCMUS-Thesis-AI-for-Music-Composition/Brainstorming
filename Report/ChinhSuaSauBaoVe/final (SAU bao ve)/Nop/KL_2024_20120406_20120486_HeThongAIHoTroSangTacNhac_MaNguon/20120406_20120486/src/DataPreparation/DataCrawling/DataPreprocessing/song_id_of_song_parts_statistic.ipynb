```python
import os
from os.path import join, getsize
import re

import sys
sys.path.append(root_directory)
import random

from bs4 import BeautifulSoup

import file_utils as fu
```

```python
song_ids_of_song_parts_dir = fu.relative_to_absolute_path(
    "DataPreprocessing/song_id_of_song_parts",
    root_path=root_directory
)

print(song_ids_of_song_parts_dir)
```

```python
song_ids_quantity_frequency = {}

start_letter = "a"
end_letter = "a"

for letter in range(ord(start_letter), ord(end_letter) + 1):
    letter_folder_path = fu.relative_to_absolute_path(
        f"{chr(letter)}",
        root_path=song_ids_of_song_parts_dir
    )

    if not os.path.exists(letter_folder_path):
        break

    dirs = os.listdir(letter_folder_path)

    for dir in dirs:
        # Lấy danh sách các file trong thư mục dir
        absolute_dir_path = fu.relative_to_absolute_path(dir, root_path=letter_folder_path)

        file_list =[ 
            fu.relative_to_absolute_path(
                file, 
                root_path=absolute_dir_path
            ) for file in os.listdir(fu.relative_to_absolute_path(
                dir,
                root_path=letter_folder_path
            ))
        ]

        for file_path in file_list:
            song_parts_ids = fu.read_data_from_json_file(file_path)["song_parts_ids"]
            count_id = len(song_parts_ids)

            print(f"----------> {count_id} parts: {file_path}")
            
            if count_id not in song_ids_quantity_frequency:
                song_ids_quantity_frequency[count_id] = 1
            else:
                song_ids_quantity_frequency[count_id] += 1

# Sort dict key by ascending order
song_ids_quantity_frequency = dict(sorted(song_ids_quantity_frequency.items()))

print(song_ids_quantity_frequency)
```
