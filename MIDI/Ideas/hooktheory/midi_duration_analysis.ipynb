```python
from midi_file_utils import read_midi_file

import os
from mido import MidiFile
```

```python
# List all files in the directory
files = os.listdir(midi_dir)

print(len(files))
```

```python
midi_duration_in_second_range_quantity_dict = {}
```

```python
for i in range(0, 180, 5):
    min_dur = i
    max_dur = i + 5
    
    key = f"{min_dur} <= duration < {max_dur}"

    midi_duration_in_second_range_quantity_dict[key] = {
        "min_duration": min_dur,
        "max_duration": max_dur,
        "count": 0
    }

midi_duration_in_second_range_quantity_dict[
    "duration >= 180"
] = {
    "min_duration": 180,
    "max_duration": 100 * 60, # 100 minutes
    "count": 0
}

print(midi_duration_in_second_range_quantity_dict)
```

```python
current_file_index = 0

for file in files:
    if "DS_Store" in file:
        continue
    else:
        pass
    
    print(f"PROCESSING {current_file_index}/{len(files)}: {file}")

    file_path = os.path.join(midi_dir, file)

    mid = MidiFile(file_path)

    duration = mid.length

    for key in midi_duration_in_second_range_quantity_dict:
        range_dict = midi_duration_in_second_range_quantity_dict[key]

        if range_dict["min_duration"] <= duration < range_dict["max_duration"]:
            range_dict["count"] += 1

    current_file_index += 1
```

```python
sorted_duration_quantities: list = sorted(
    midi_duration_in_second_range_quantity_dict.items(),
    key=lambda x: x[1]["count"],
    reverse=True
)

sorted_duration_quantities: list = [
    (k, v["count"]) 
    for k, v in sorted_duration_quantities
]

# Make the key beautiful to display (each key has the same length, and the duration string is all aligned)
sorted_duration_quantities: list = [
    (k.ljust(30, " "), v) 
    for k, v in sorted_duration_quantities
]

for k, v in sorted_duration_quantities:
    print(f"{k}: {str(v).ljust(4, ' ')} file(s)")
```
