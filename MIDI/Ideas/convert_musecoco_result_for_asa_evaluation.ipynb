```python
import sys

sys.path.append(midi_lib_path)

import json
import os

import converter.musecoco_line_str_to_midi  as text_to_midi
```

```python
# List all txt files with its name match 0.txt, 1.txt, 2.txt, ... (recursively)
file_paths: list[str] = []

accepted_file_names: list[int] = [number for number in range(0, 10)]

for root, dirs, files in os.walk(top_k_dir):
    for file in files:
        if file.endswith(".txt"):
            absoulte_path = os.path.join(root, file)
            file_name = os.path.basename(absoulte_path)
            file_name_without_ext = os.path.splitext(file_name)[0]
            
            if file_name_without_ext.isnumeric() and int(file_name_without_ext) in accepted_file_names:
                file_paths.append(absoulte_path)
            else:
                pass

for file in file_paths:
    print(file)
```

```python
# Extract data from txt files to its corresponding midi directory
for file in file_paths:
    txt_file_grand_parent_dir = os.path.dirname(os.path.dirname(file))
    corresponding_midi_dir = os.path.join(txt_file_grand_parent_dir, "midi")

    file_name_without_ext = os.path.splitext(os.path.basename(file))[0]
    
    with open(file, "r") as f:
        text_data = f.read()
        text_data = text_data.split("<sep>")[1].strip()

        text_to_midi.set_default_values(
            output_path=corresponding_midi_dir,
            log_path=txt_file_grand_parent_dir,
            midi_file_name_prefix=""
        )

        (
            midi_file_path, 
            original_structure_log_path, 
            wrong_structure_log_path
        ) = text_to_midi.musecoco_line_str_to_midi_converter(
            text_data,
            should_fill_missing = False
        )

        new_file_name = os.path.join(
            corresponding_midi_dir, 
            f"{file_name_without_ext}.mid"
        )
        
        os.rename(midi_file_path, new_file_name)

        print(f"Converted {file}")
        print(f"-----> midi file: {new_file_name}")
        print(f"-----> original structure log: {original_structure_log_path}")
        print(f"-----> wrong structure log: {wrong_structure_log_path}")
        print("")
```
