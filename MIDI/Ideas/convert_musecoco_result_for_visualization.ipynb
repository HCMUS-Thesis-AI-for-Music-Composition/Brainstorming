# Data concatenation

```python
data_part_1_dir = f"{model_text_result_dir}/part 1"
data_part_2_dir = f"{model_text_result_dir}/part 2"
```

```python
part_1_txt_file_path = f"{data_part_1_dir}/muse_output.txt"

part_1_txt_data: list[str] = []

with open(part_1_txt_file_path, "r") as f:
    part_1_txt_data = f.read().replace(
        "[", ""
    ).replace(
        "]", ""
    ).replace(
        "\n", ""
    ).replace(
        "'", ""
    ).split(", ")

print(f"Found {len(part_1_txt_data)} results in part 1")
print(part_1_txt_data)
```

```python
part_2_txt_files: list[str] = []
part_2_txt_data: list[str] = []

# List all txt files in part 2 (recursively)
for root, dirs, files in os.walk(data_part_2_dir):
    for file in files:
        absoulte_path = os.path.join(root, file)
        if file.endswith(".txt") and "remi" in absoulte_path:
            part_2_txt_files.append(os.path.join(root, file))

# Read all txt files in part 2
for file in part_2_txt_files:
    with open(file, "r") as f:
        part_2_txt_data.append(f.read())

# Remove empty strings
part_2_txt_data = list(filter(None, part_2_txt_data))

print(f"Found {len(part_2_txt_files)} results in part 2")
print(part_2_txt_data)
```

```python
total_result_data: list[str] = []
total_result_data.extend(part_1_txt_data)
total_result_data.extend(part_2_txt_data)

print(f"Total results: {len(total_result_data)}")
print(total_result_data)

# Save to json
json_file_path = f"{model_text_result_dir}/total_result.json"

with open(json_file_path, "w") as f:
    json.dump(total_result_data, f)

print(f"Saved to {json_file_path}")
```
