```python
import os
import shutil
```

# Infer Command

```python
# List all infer_command files in musecoco_top_k_dir (recursively)
musecoco_infer_command_files = []

command_filename = "infer_command.json"

for root, dirs, files in os.walk(musecoco_top_k_dir):
    for file in files:
        if command_filename == file[-len(command_filename):]:
            absolute_path = os.path.join(root, file)
            musecoco_infer_command_files.append(absolute_path)

print("Number of infer_command files in musecoco_top_k_dir:", len(musecoco_infer_command_files))
for file in musecoco_infer_command_files[:5]:
    print(file)
```

```python
# Read all infer_command files in musecoco_top_k_dir
musecoco_infer_commands: list[dict] = []

for file in musecoco_infer_command_files:
    with open(file, "r", encoding='UTF-8') as f:
        command = f.read()
        command = command.replace("\u2014", "-")
        musecoco_infer_commands.append({
            "file": file,
            "infer_command": command
        })

# Sort infer_commands by file name
musecoco_infer_commands.sort(key=lambda x: x["file"])

for command in musecoco_infer_commands[:5]:
    print(command["file"])
    print(command["infer_command"])
    print()
```

# Prompt

```python
# List all prompt files in gpt_top_k_dir (recursively)
gpt_prompt_files = []

prompt_filename = "prompt.txt"

for root, dirs, files in os.walk(gpt_top_k_dir):
    for file in files:
        if prompt_filename == file[-len(prompt_filename):]:
            absolute_path = os.path.join(root, file)
            gpt_prompt_files.append(absolute_path)

print("Number of prompt files in gpt_top_k_dir:", len(gpt_prompt_files))
for file in gpt_prompt_files[:5]:
    print(file)
```

```python
# Read all prompt files in gpt_top_k_dir
prompts: list[dict] = []

for file in gpt_prompt_files:
    with open(file, "r", encoding='UTF-8') as f:
        prompt = f.read()

        # Fix utf-8 encoding issue
        prompt = prompt.replace("\u2014", "-")
        
        prompts.append({
            "file": file,
            "content": prompt
        })

# Sort gpt_infer_commands by file name
prompts.sort(key=lambda x: x["file"])

for prompt in prompts[:5]:
    print(prompt["file"])
    print(prompt["content"])
    print()
```

# Match and Copy

```python
n_matching = 0

for prompt in prompts:
    for command in musecoco_infer_commands:
        parent_folder_of_prompt = os.path.dirname(prompt["file"])
        parent_folder_of_infer_command = os.path.dirname(prompt["file"])

        if prompt["content"][2:-2] in command["infer_command"]:
            n_matching += 1
    
            # print(f'-----> Prompt file: {prompt["file"]}')
            # print(f'-----> Infer Command: {infer_command["file"]}')

            # print("-----> Parent folder of prompt:", parent_folder_of_prompt)
            
            # Copy the infer_command file to the parent folder of the prompt file
            infer_command_file_name = "infer_command.json"
            infer_command_file_destination = os.path.join(parent_folder_of_prompt, infer_command_file_name)

            shutil.copy(prompt["file"], infer_command_file_destination)
            print("MATCHING: -----> Copied to:", infer_command_file_destination)
            print()
        else:
            _parent_folder_of_infer_command = int(parent_folder_of_infer_command.split("/")[-1])
            _parent_folder_of_prompt = int(parent_folder_of_prompt.split("/")[-1])
            
            if (
                # (_parent_folder_of_infer_command == _parent_folder_of_prompt) 
                # (_parent_folder_of_infer_command == _parent_folder_of_prompt + 1)
                (_parent_folder_of_infer_command == _parent_folder_of_prompt - 1)
            ):
                print("NOT MATCHING")
                print(f'-----> Prompt file: {prompt["file"]}')
                print(f'-----> Infer Command: {prompt["file"]}')
                print("-----> Prompt content:", prompt["content"])
                print("-----> Infer Command content:", prompt["infer_command"])
                print()

print("Number of matching:", n_matching)
```
