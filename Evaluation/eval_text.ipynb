```python
def listing_txt_and_json(
    folder_path: str,
    recursively: bool = True
):
    """
    Listing all .txt and .json files in the folder
    """
    txt_files = []
    json_files = []
    
    for root, _, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".txt"):
                txt_files.append(os.path.join(root, file))
            if file.endswith(".json"):
                json_files.append(os.path.join(root, file))
        
        if not recursively:
            break
        
    result = {
        "txt": txt_files,
        "json": json_files
    }

    return result
```
