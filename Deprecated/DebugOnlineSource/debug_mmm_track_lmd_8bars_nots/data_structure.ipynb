```python
default_n_displaying_characters = 18
def adding_trailing_space(
    text, 
    n_displaying_characters=default_n_displaying_characters
):
    n_spaces_to_add = n_displaying_characters - len(text) - 1

    if n_spaces_to_add < 0:
        return text[:(n_displaying_characters - 1)]
    else:
        return text + (' ' * n_spaces_to_add)
```
