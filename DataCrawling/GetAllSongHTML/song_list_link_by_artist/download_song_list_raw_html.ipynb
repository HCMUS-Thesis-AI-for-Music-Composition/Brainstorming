```python
import sys
sys.path.append(root_directory)

import os
from os.path import join, getsize
import time

import requests

import file_utils as fu
```

```python
artist_links = fu.read_data_from_json_file(
    fu.relative_to_absolute_path(
        "GetAllSongHTML/artist_list_link_by_alphabet/artist_links.json",
        root_directory
    )        
)

print(artist_links)
```

```python
raw_html_folder = fu.relative_to_absolute_path(
            f"GetAllSongHTML/song_list_link_by_artist/raw_html", 
            root_path=root_directory
        )

# Tạo các folder từ a đến z trong thư mục raw_html
for i in range(26):
    # Kiểm tra xem folder đã tồn tại chưa
    if not os.path.exists(join(raw_html_folder, chr(ord('a') + i))):
        os.mkdir(join(raw_html_folder, chr(ord('a') + i)))
```

```python
def get_html_content(url, file_path):
    print(f"Getting {url}...")
    response = requests.get(url)
    with open(file_path, "w", encoding='utf-8') as f:
        f.write(response.text)
```

```python
def get_song_list_link_raw_html_content(from_letter='a'):
    for link in artist_links:
        # Lấy các ký tự sau dấu / cuối cùng trong link làm tên file
        file_name = link["link"].split("/")[-1]
        letter = file_name[0].lower()

        if letter < from_letter.lower():
            continue
        else:
            get_html_content(
                link["link"], 
                fu.relative_to_absolute_path(
                    f"GetAllSongHTML/song_list_link_by_artist/raw_html/{letter}/{file_name}.html", 
                    root_path=root_directory
                )
            )

            time.sleep(0.95) # Rate limiting
```
