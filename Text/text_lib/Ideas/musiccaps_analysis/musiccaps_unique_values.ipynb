```python
inst_class_id_to_inst_class_name = {
    # piano 0:
    0: ['piano', "grand piano"],

    # keyboard 1:
    1: ['keyboard', "digital keyboard", "synthesizer keyboard"],

    # percussion 2:
    2: ['percussion'],

    # organ 3:
    3: ['organ'],

    # guitar 4:
    4: ['guitar'],

    # bass 5:
    5: ['bass'],

    # violin 6:
    6: ['violin'],

    # viola 7:
    7: ['viola'],

    # cello 8:
    8: ['cello'],

    # harp 9:
    9: ['harp'],

    # strings 10:
    10: ['strings'],

    # voice 11:
    11: ['voice'],

    # trumpet 12:
    12: ['trumpet'],

    # trombone 13:
    13: ['trombone'],

    # tuba 14:
    14: ['tuba'],

    # horn 15:
    15: ['horn'],

    # brass 16:
    16: ['brass'],

    # sax 17:
    17: ['sax', "saxophone"],

    # oboe 18:
    18: ['oboe'],

    # bassoon 19:
    19: ['bassoon'],

    # clarinet 20:
    20: ['clarinet'],

    # piccolo 21:
    21: ['piccolo'],

    # flute 22:
    22: ['flute'],

    # pipe 23:
    23: ['pipe'],

    # synthesizer 24:
    24: ['synthesizer', "synth"],

    # ethnic instrument 25:
    25: ['ethnic instrument'],

    # sound effect 26:
    26: ['sound effect'],

    # drum 27:
    27: ['drum'],
}
```

```python
# Convert csv to dict
import pandas as pd

df = pd.read_csv(csv_file_path)

df.head()
```

```python
aspect_list= df["aspect_list"]

aspect_list.head()
```

```python
def replace_by_mapper(text, mapper: dict):
    for key, value in mapper.items():
        text = text.replace(key, value)
    
    return text
```

```python
aspect_list = aspect_list.apply(
    lambda x: replace_by_mapper(
        x, 
        {
            "['": "",
            "']": "",
            "', '": ", ",
            " ": "",
            "vocal": "voice"
        }
    ).split(",")
)

aspect_list.head()
```

```python
# aspect_list to set
aspect_set = set()
for aspect in aspect_list:
    aspect_set.update(aspect)

print(len(aspect_set))

aspect_set
```

```python
aspect_list = list(aspect_set)

print(aspect_list)
```

```python
supported_instrs = [
    instr.replace(" ", "").lower()
    for idx, instrs in inst_class_id_to_inst_class_name.items()
    for instr in instrs
]

print(supported_instrs)
```

```python
aspects_contain_supported_instrs: dict = {
    instr: []
    for instr in supported_instrs
}

for instr in supported_instrs:
    for aspect in aspect_list:
        if instr in aspect.lower():
            aspects_contain_supported_instrs[instr].append(aspect)
```

```python
for k, v in aspects_contain_supported_instrs.items():
    print(f"{k}:\n    n = {len(v)}\n    aspects: {v}\n")
```

```python
n_aspects_contain_supported_instrs = sum(
    len(v) for v in aspects_contain_supported_instrs.values()
)

print(f"n_aspects_contain_supported_instrs = {n_aspects_contain_supported_instrs}")
```
